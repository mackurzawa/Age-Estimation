```python
import os
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
```

```python
FILENAMES = os.listdir(os.path.join('Data', 'UTKFace'))
FILENAMES[:5]
```

```python
ages = [int(x.split('_')[0]) for x in FILENAMES]
age_distribution = {}
for i in range(min(ages), max(ages)+1):
    age_distribution[i] = 0
    
for age in ages:
    age_distribution[age] += 1

plt.figure(figsize=(10, 7))
plt.bar(age_distribution.keys(), age_distribution.values(), width=1)
plt.title('Age distribution')
plt.xlabel('Age')
plt.ylabel('Quantity')
plt.show()
```

```python
AGE_THRESHOLDS = [2, 7, 14, 20, 25, 29, 35, 40, 48, 55, 65]
AGE_THRESHOLDS.insert(0, min(ages)-1)
AGE_THRESHOLDS.append(max(ages))

classes_names = []
classes_middle = []
age_classes_distribution = {}
for i_thresh, thresh in enumerate(AGE_THRESHOLDS[:-1]):
    class_name = f'{thresh+1} - {AGE_THRESHOLDS[i_thresh + 1]}'
    classes_names.append(class_name)
    classes_middle.append((thresh + 1 + AGE_THRESHOLDS[i_thresh + 1])/2)
    age_classes_distribution[class_name] = 0
    
print(classes_names)
print(classes_middle)
    
i = 0

for age in age_distribution:
    if age <= AGE_THRESHOLDS[i+1]:
        age_classes_distribution[classes_names[i]] += age_distribution[age]
    else:
        i += 1
        
plt.figure(figsize=(10, 7))
plt.bar(age_classes_distribution.keys(), age_classes_distribution.values())
```

# LEARNING

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
from torchvision.io import read_image
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
class AgeImageDataset(Dataset):
    def __init__(self, images_dir=os.path.join('Data', 'Train')):
        self.images_dir = images_dir
        self.images_names = os.listdir(self.images_dir)
    
    def __len__(self):
        return len(self.images_names)
    
    def __getitem__(self, idx):
        image_name = self.images_names[idx]
        image = read_image(os.path.join(self.images_dir, image_name)) / 255.
        label_temp = image_name.split('_')[0]
        for i in range(len(AGE_THRESHOLDS) - 1):
            if AGE_THRESHOLDS[i] < int(label_temp) <= AGE_THRESHOLDS[i+1]:
                label = classes_middle[i]
                label = i
                break
        return image.float(), label
```

```python
dataset = AgeImageDataset()
dataset = AgeImageDataset(os.path.join('Data', 'Test'))
train_dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
def show_random_photo():
    data = next(iter(train_dataloader))
    photo = data[0][0]
    label = data[1][0]
    plt.imshow(np.transpose(photo, (1, 2, 0)))
    plt.title(f'Label: {label}')
#     print(label)

show_random_photo()
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.l1 = nn.Conv2d(3, 8, 5)
        self.l2 = nn.Conv2d(8, 64, 5)
        self.l3 = nn.Conv2d(64, 128, 5)
        self.l4 = nn.Conv2d(128, 256, 5)
        self.l5 = nn.Conv2d(256, 256, 5)
        self.l6 = nn.Linear(1024, 128)
        self.l7 = nn.Linear(128, 10)
        self.l8 = nn.Linear(10, 1)
        
    def forward(self, x):
        x = F.relu(self.l1(x))
        x = F.max_pool2d(x, 2)
        x = F.relu(self.l2(x))
        x = F.max_pool2d(x, 2)
        x = F.relu(self.l3(x))
        x = F.max_pool2d(x, 2)
        x = F.relu(self.l4(x))
        x = F.max_pool2d(x, 2)
        x = F.relu(self.l5(x))
        x = F.max_pool2d(x, 2)
        x = torch.flatten(x, 1)
        x = F.relu(self.l6(x))
        x = F.relu(self.l7(x))
        x = self.l8(x)
                
        return x.double()

model = Model()
print(model(torch.randn(1, 3, 200, 200)).shape)
```

```python
class Model_2(nn.Module):
    def __init__(self):
        super(Model_2, self).__init__()
        self.l1 = nn.Conv2d(3, 8, 5)
        self.l2 = nn.Conv2d(8, 64, 5)
        self.l3 = nn.Conv2d(64, 128, 5)
        self.l4 = nn.Conv2d(128, 256, 5)
        self.l5 = nn.Conv2d(256, 256, 5)
        self.l6 = nn.Linear(1024, 128)
        self.l7 = nn.Linear(128, len(classes_names))
        
    def forward(self, x):
        x = F.relu(self.l1(x))
        x = F.max_pool2d(x, 2)
        x = F.relu(self.l2(x))
        x = F.max_pool2d(x, 2)
        x = F.relu(self.l3(x))
        x = F.max_pool2d(x, 2)
        x = F.relu(self.l4(x))
        x = F.max_pool2d(x, 2)
        x = F.relu(self.l5(x))
        x = F.max_pool2d(x, 2)
        x = torch.flatten(x, 1)
        x = F.relu(self.l6(x))
        x = self.l7(x)
                
        return x.double()

model_2 = Model_2()
model_2.to(device)
print(model_2(torch.randn(1, 3, 200, 200).to(device)).shape)
```

```python
EPOCHS = 10
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = optim.Adam(model_2.parameters())

for epoch in range(EPOCHS):
    losses = []
    for X, y in tqdm(train_dataloader):
        optimizer.zero_grad()
        
        pred = model_2(X.to(device))
        loss = loss_fn(pred, y.to(device))
        loss.backward()
        optimizer.step()
        
#         print(y.shape, pred.argmax().shape)
        
        losses.append(loss)
    print(f'Epoch {epoch+1} completed')
    print(f'Loss: {sum(losses)/len(losses)}')
    print()
```

```python
def predict_random_photo():
    data = next(iter(train_dataloader))
    photo = data[0][0]
    label = data[1][0]
    plt.imshow(np.transpose(photo, (1, 2, 0)))
    plt.title(f'Label: {label}, Predicted: {torch.argmax(model_2(torch.Tensor([photo.numpy()]).to(device))[0])}')
#     print(label)

predict_random_photo()
```
